```python
import pandas as pd
import numpy as np
import os
import time
```

```python
data_folders = ['GSM3511735', 'GSM3511747', 'GSM3511752', 'GSM3511757', 'GSM3511762',
                'GSM3511766', 'GSM3721449', 'GSM3721453', 'GSM3721454', 'GSM3721455', 
                'GSM3721457', 'GSM3721458', 'GSM3721460', 'GSM3721462', 
                'GSM3721464',
               
                'GSM3511738', 'GSM3511741', 'GSM3511749', 'GSM3511754', 'GSM3511760', 
                'GSM3721445', 'GSM3721447', 'GSM3721448', 'GSM3721450', 'GSM3721451', 
                'GSM3721452', 'GSM3721456', 'GSM3721459', 'GSM3721461', 'GSM3721463', 
                'GSM3892576',
                
                'GSM3892571', 'GSM3892570'               
               ]
```

```python
folder_path = os.getcwd()
common_genes = set([line.rstrip('\n') for line in open(folder_path + "/genes_intersection.txt")])
print(len(common_genes))
```

```python
%%time

this_dir = os.getcwd()
print(f'U preseku ima {len(common_genes)} gena.')
print('-----------------------------------------------------------------------')
print('-----------------------------------------------------------------------')
for folder in data_folders:     
    file_names = os.listdir(os.path.join(folder_path, folder))
    csv_path = ""
    
    for file in file_names:
        if(file[0:13] == "preprocessed1"):
            csv_path = os.path.join(os.path.join(folder_path, folder), file)
            
            raw_data = pd.read_csv(csv_path)
            file_genes = set(list(raw_data.columns)[1:])
            for_dropping = list(file_genes.difference(file_genes.intersection(common_genes)))
            raw_data_drop_cols = raw_data.drop(for_dropping, axis=1)   
            
            csv_path = os.path.join(os.path.join(this_dir, folder), 
                            "preprocessed2_" + folder + ".csv")
            ### Čuvam preprocesiran fajl
            raw_data_drop_cols.to_csv(csv_path, index=False)
            
            txt_path = os.path.join(os.path.join(this_dir, folder), "difference_genes_" + folder + ".txt")
            ### Čuvam spisak gena koji nisu u preseku svih
            with open(txt_path, 'w') as output:
                for row in for_dropping:
                    output.write(str(row) + '\n')
                    
            print('-----------------------------------------------------------------------')
            print(f'Folder {folder} ima {len(for_dropping)} od ukupno {len(list(raw_data.columns)[1:])} gena koji nisu u preseku.')
            print('-----------------------------------------------------------------------')
            print()
            print()
```
