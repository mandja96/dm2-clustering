```python
import pandas as pd
import numpy as np
import os
import time
```

```python
data_folders = ['GSM3511735', 'GSM3511747', 'GSM3511752', 'GSM3511757', 'GSM3511762',
                'GSM3511766', 'GSM3721449', 'GSM3721453', 'GSM3721454', 'GSM3721455', 
                'GSM3721457', 'GSM3721458', 'GSM3721460', 'GSM3721462', 
                'GSM3721464',
               
                'GSM3511738', 'GSM3511741', 'GSM3511749', 'GSM3511754', 'GSM3511760', 
                'GSM3721445', 'GSM3721447', 'GSM3721448', 'GSM3721450', 'GSM3721451', 
                'GSM3721452', 'GSM3721456', 'GSM3721459', 'GSM3721461', 'GSM3721463', 
                'GSM3892576',
                
                'GSM3892571', 'GSM3892570'               
               ]
```

```python
%%time

this_dir = os.getcwd()

for folder in data_folders:     
    file_names = os.listdir(os.path.join(this_dir, folder))
    csv_path = ""
    
    for file in file_names:
        if(file[0:13] == "preprocessed2"):
            csv_path = os.path.join(os.path.join(this_dir, folder), file)
            
            raw_data = pd.read_csv(csv_path)
            raw_data_sorted = raw_data.sort_index(axis=1) 
            index_col = raw_data_sorted['Index']
            raw_data_sorted.drop(labels=['Index'], axis=1, inplace = True)
            raw_data_sorted.insert(0, 'Index', index_col)
            
            csv_path = os.path.join(os.path.join(this_dir, folder), 
                            "sorted_without_percent_" + folder + ".csv")
            ### Čuvam preprocesiran fajl
            raw_data_sorted.to_csv(csv_path, index=False)
```
